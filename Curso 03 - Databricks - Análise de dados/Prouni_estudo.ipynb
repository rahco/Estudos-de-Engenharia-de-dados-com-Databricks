```
# Buscando via sql as colunas bolsa_integral_cotas e uf_busca do df
ps.sql('''
SELECT bolsa_integral_cotas, uf_busca
FROM {DF}
''', DF=df)
```

```
# Filtrando os cursos de medicina
ps.sql('''
SELECT bolsa_integral_cotas, uf_busca
FROM {DF}
WHERE nome_curso = 'Medicina'
LIMIT 5
''', DF=df)
```

```
# Gerando df com a soma de bolsa_integral_cotas por uf
cotas_ordenadas = ps.sql('''
SELECT uf_busca, SUM(bolsa_integral_cotas) AS Total_de_cotas
FROM {DF}
WHERE nome_curso='Medicina'
GROUP BY uf_busca
ORDER BY Total_de_cotas DESC
''', DF=df)
```

```
cotas_ordenadas.head()
```

```
# Plotando o resultado do df em um gráfico de barras
cotas_ordenadas.plot.bar(x='uf_busca',y='Total_de_cotas')
```

```
# Verificando cidades únicas
df_medicina['cidade_busca'].unique()
```

```
# Criando um dicionário com a relação das cidades e se são ou não capitais
nomes_cidades = {
"Sao Luis": "Capital",
"Joao Pessoa":"Capital",
"Belem":"Capital",
"Itaperuna":"Município",
"Vitoria":"Município",
"Franca":"Município",
"Sao Paulo":"Capital",
"Guarapuava":"Município",
"Campo Mourao":"Município",
"Montes Claros":"Município",
"Rio Branco":"Município",
"Imperatriz":"Município",
"Vespasiano":"Município",
"Porto Velho": "Município"
}
```

```
# Adicionando a relação de classificação das cidades em uma variável com o método .map
capitais = df_medicina['cidade_busca'].map(nomes_cidades)
```

```
# Utilizando o método .insert para adicionar a coluna tipo_cidade ao df (o parâmetro loc define a posição da coluna no df)
df_medicina.insert(loc=8,column='tipo_cidade',value=capitais)
```

```
df_medicina.head()
```

```
# Gerando a consulta e plotagem do total de cotas por tipo_cidade
cotas_ordenadas = ps.sql('''
SELECT tipo_cidade, SUM(bolsa_integral_cotas) AS Total_de_cotas
FROM {DF}
GROUP BY tipo_cidade
ORDER BY Total_de_cotas DESC
''', DF=df_medicina)
cotas_ordenadas.plot.bar(x='tipo_cidade',y='Total_de_cotas')
```

Verificando no dataset quais faculdades são universidades.

```
# Utilizando o método de string .str filtrando com o .contains nomes que possuem Universidade
selecao = df_medicina[df_medicina['universidade_nome'].str.contains('Universidade')][['universidade_nome']]
```

```
selecao
```

```
# Tratando/substituindo no df o nome da única universidade que não possui o padrão - SIGLA
df_medicina['universidade_nome'].loc[42] = 'Centro Universitário Integrado de Campo Mourão - CUICM'
df_medicina['universidade_nome'].loc[42]
```

```
# Separando o df com o método .split utilizando como parâmetro ' - '
separado = df_medicina['universidade_nome'].str.split(' - ',n=1,expand=True)
separado
```

```
# Adicionando a coluna criada no df separado ao df_medicina
df_medicina.insert(loc=12,column='sigla',value=separado[1])
```

```
df_medicina.head()
```

```
# Plotando o gráfico com a média de nota_integral_ampla por sigla
notas_ordenadas = ps.sql('''
SELECT sigla, MEAN(nota_integral_ampla) AS nota_media
FROM {DF}
GROUP BY sigla
ORDER BY nota_media DESC
''', DF=df_medicina)
notas_ordenadas.plot.bar(x='sigla',y='nota_media')
```
